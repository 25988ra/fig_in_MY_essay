````python
import numpy as np 
import matplotlib.pyplot as plt 
import pandas as pd 
import xlrd 


# 获得插值公式

# x = [float(i) for i in input("plz input x elements").split()] # 这里的元素对接excel的中储存的元素
# y = [float(i) for i in input("plz input y elements").split()]
x = [0,1,2]
y = [3,-2,1]
x_iter = np.array(x)
y_iter = np.array(y)
delta_x =  np.array([x[i+1]-x[i] for i in range(0,len(x)-1)])
delta_y =  np.array([y[i+1]-y[i] for i in range(0,len(y)-1)])
r = np.zeros(len(x))
matrix_delta = np.diag(np.ones(len(x)))
#这里最好对b转置一下
for i in range(1,len(x)-1):
    r[i] += 3*(delta_y[i]/delta_x[i] - delta_y[i-1]/delta_x[i-1])
    matrix_delta[i][i-1] += delta_x[i-1]
    matrix_delta[i][i] += 2*delta_x[i-1]+2*delta_x[i] - 1
    matrix_delta[i][i+1] += delta_x[i] 
c = np.dot(np.linalg.inv(matrix_delta),r.reshape(len(r),1))
d = np.array(
    [
        (c[i+1] - c[i])/3*delta_x[i] for i in range(len(c)-1)
    ])
b = np.array(
    [
        (delta_y[i]/delta_x[i] - (2*c[i] + c[i+1])*delta_x[i]/3) for i in range(len(c)-1)
    ]
    )
# 得到方程表达式
def s1(x):
    delta = (x-x_iter[0])
    return y_iter[0]+b[0]*delta + c[0] * delta**2 + d[0] *delta**3
def s2(x):
    delta = (x-x_iter[1])
    return y_iter[1]+b[1]*delta + c[1] * delta**2 + d[1] *delta**3

# 设置区间
x_step = np.linspace(-0.5,2.5,100)
interval0 = [1 if (i<1) else 0 for i in x_step]
interval1 = [1 if (i>=1) else 0 for i in x_step]

y_function = s1(x_step)*interval0  + s2(x_step)*interval1

# 绘制图片 
plt.scatter(x_iter,y_iter,c='red')
plt.plot(x_step,y_function)
plt.show()



````
